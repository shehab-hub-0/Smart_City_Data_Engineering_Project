```python
import os
import sys
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, split, regexp_replace

# --- 1. Configuration ---
SILVER_TABLE = "nessie.silver_city_events"
GOLD_DB = "nessie"

MINIO_ENDPOINT = "http://minio:9000"
MINIO_ACCESS_KEY = os.getenv("MINIO_ROOT_USER", "minioadmin")
MINIO_SECRET_KEY = os.getenv("MINIO_ROOT_PASSWORD", "minioadmin")
NESSIE_URL = "http://nessie:19120/api/v1"

def build_spark():
    return SparkSession.builder \
        .appName("SmartCity_Gold_Final") \
        .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions,org.projectnessie.nessie-integrations.nessie-spark-extensions-3.5_2.12") \
        .config("spark.sql.catalog.nessie", "org.apache.iceberg.spark.SparkCatalog") \
        .config("spark.sql.catalog.nessie.uri", NESSIE_URL) \
        .config("spark.sql.catalog.nessie.ref", "main") \
        .config("spark.sql.catalog.nessie.catalog-impl", "org.apache.iceberg.nessie.NessieCatalog") \
        .config("spark.sql.catalog.nessie.io-impl", "org.apache.iceberg.hadoop.HadoopFileIO") \
        .config("spark.sql.catalog.nessie.warehouse", "s3a://warehouse") \
        .config("spark.hadoop.fs.s3a.endpoint", MINIO_ENDPOINT) \
        .config("spark.hadoop.fs.s3a.access.key", MINIO_ACCESS_KEY) \
        .config("spark.hadoop.fs.s3a.secret.key", MINIO_SECRET_KEY) \
        .config("spark.hadoop.fs.s3a.path.style.access", "true") \
        .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
        .config("spark.sql.defaultCatalog", "nessie") \
        .getOrCreate()

def main():
    spark = build_spark()
    spark.sparkContext.setLogLevel("WARN")

    # 1. READ FROM SILVER
    silver_df = spark.read.format("iceberg").load(SILVER_TABLE)

    # 2. PRE-PROCESSING
    processed_df = silver_df \
        .withColumn("loc_clean", regexp_replace(col("location"), "[\(\)]", "")) \
        .withColumn("latitude", split(col("loc_clean"), ",").getItem(0).cast("double")) \
        .withColumn("longitude", split(col("loc_clean"), ",").getItem(1).cast("double")) \
        .withColumn("traffic_volume", col("traffic_volume").cast("integer")) \
        .withColumn("average_speed", col("average_speed").cast("double")) \
        .withColumn("temperature", col("temperature").cast("double")) \
        .withColumn("air_quality_index", col("air_quality_index").cast("integer")) \
        .withColumn("pm25", col("pm25").cast("double"))

    # 3. CREATE/REPLACE DIMENSIONS (Using writeTo API)
    # Dim Zones
    dim_zones = processed_df.select("city_zone", "latitude", "longitude").distinct()
    dim_zones.writeTo(f"{GOLD_DB}.dim_zones").createOrReplace()
    print(f"✅ Created/Updated: {GOLD_DB}.dim_zones")

    # Dim Emergency Types
    dim_emergency = processed_df.select("emergency_type", "priority_level").distinct()
    dim_emergency.writeTo(f"{GOLD_DB}.dim_emergency_types").createOrReplace()
    print(f"✅ Created/Updated: {GOLD_DB}.dim_emergency_types")

    # 4. CREATE/REPLACE FACT TABLE
    fact_city_events = processed_df.select(
        "id", "timestamp", "city_zone", "domain", "device_id",
        "traffic_volume", "average_speed", "temperature", "weather_condition",
        "air_quality_index", "pm25", "emergency_type", "emergency_status"
    )
    fact_city_events.writeTo(f"{GOLD_DB}.fact_city_events").createOrReplace()

    print(f"🏆 GOLD STAR SCHEMA COMPLETED!")

if __name__ == "__main__":
    main()
```
